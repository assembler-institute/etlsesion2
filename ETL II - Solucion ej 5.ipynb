```python
# First is needed to install tweepy

# pip install tweepy

import tweepy
import pandas as pd

access_token = ''
access_token_secret = ''
api_key = ''
api_key_secret = ''
bearer_token = ''
```

```python
client = tweepy.Client(
    bearer_token = bearer_token,
    consumer_key = api_key,
    consumer_secret = api_key_secret,
    access_token = access_token,
    access_token_secret = access_token_secret
)
```

```python
searchs = ['@iberia', '@Fnac_ESP', '@Renfe', '#AVE', '#ClubFnac', '#VolverAVolar', '#Equipaje']
```

```python
results = []

for term in searchs:
    search = client.search_recent_tweets(query=term, tweet_fields=['author_id', 'created_at', 'source' ] ,max_results=100)
    
    
    try:
        for tweet in search.data:
    
            template = {
                "id": tweet.id,
                "date": tweet.id,
                "query": term,
                "content": tweet.text
            }
        
            results.append(template)
    except:
        pass
    
df = pd.DataFrame(results, columns=results[0].keys())
```
